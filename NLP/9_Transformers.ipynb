````python
from transformers import AutoModel, AutoTokenizer
import torch
````

## Modelin tanımlanması

````python
model_name = "dbmdz/bert-base-turkish-cased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
````

````python
ornek_metin = """ 
Türkiye'nin başkenti Ankara'dır. İstanbul ise en kalabalık şehridir. Doğal güzellikler acısından Kapadokya mukkale de termal süları ile meşhurdur.
Türk mutfağı dünyaca ünlüdür. Kebap, baklava ve Türk kahvesi en bilinen lezzetlerdir.
Boğaziçı Köprüsü İstanbul'un simgesidir. Galata Kulesi tarihi bir yapıdır.
"""
````

````python
tokens = tokenizer(ornek_metin, return_tensors="pt")
````

````python
def fonk(**tokens):
    for key, value in tokens.items():
        print(key, value)
fonk(**tokens)
````

````python
with torch.no_grad():
    outputs = model(**tokens)
last_hidden_states = outputs.last_hidden_state
first_token_tensor = last_hidden_states[0,0,:].numpy()
print("ilk token:")
print(first_token_tensor)
print(first_token_tensor.shape)
````
