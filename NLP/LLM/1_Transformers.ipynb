### 1. Transformer Yapısı

### Nedir Bu Transformer?

Transformer mimarisi, 2017 yılında Google’ın "Attention Is All You Need" başlıklı makalesiyle tanıtılmış devrim niteliğinde bir derin öğrenme mimarisidir. Geleneksel RNN (Recurrent Neural Network) ve LSTM modellerinin sıralı işlemeye dayalı yavaşlığı ve uzun bağımlılıkları öğrenmedeki zorluklarını aşmak için geliştirilmiştir.

Transformer, **sadece dikkat mekanizmalarına (attention)** dayanarak metin işleme yapar. Bu sayede paralelleştirme imkanı sunar ve eğitim süresini büyük ölçüde kısaltır. Bugün LLM'ler (Büyük Dil Modelleri) olarak bildiğimiz modellerin (BERT, GPT, T5, vb.) temelini bu mimari oluşturur.

### Neden Transformer Kullanılır?

- **Paralelleştirilebilirlik**: Tüm kelime dizisi aynı anda işlenebilir. RNN'lerde olduğu gibi sırayla ilerlemek gerekmez.
- **Uzun-bağlımlılık (long-range dependency)**: Kelimeler arası uzak ilişkileri kolayca yakalayabilir.
- **Esneklik**: Aynı mimari hem metin üretimi (GPT), hem sınıflandırma (BERT), hem de çeviri (T5) gibi farklı görevlerde kullanılabilir.

### Temel Bileşenler

Transformer mimarisi genel olarak iki ana bölümden oluşur:

1. **Encoder (Kodlayıcı)**: Girdi dizisini işler, bağlama (context) dayalı temsiller üretir.
2. **Decoder (Çözücü)**: Encoder çıktısını kullanarak yeni bir dizi üretir (örneğin, çeviri).

Ancak bazı modeller sadece Encoder (BERT), bazıları sadece Decoder (GPT), bazıları ise hem Encoder hem Decoder (T5) kullanır.

### Anahtar: Dikkat Mekanizması (Attention)

Transformer’ın kalbi **Self-Attention** mekanizmasıdır. Her kelime diğer tüm kelimelerle olan ilişkisini değerlendirir ve bu ilişkilerin önemini ağırlıklandırır.

Matematiksel olarak, Self-Attention şu şekilde hesaplanır:

Attention(Q,K,V)=softmax(dkQKT)V

- Q: Query (sorgu) – hangi kelimeler ilgimi çekiyor?
- K: Key (anahtar) – hangi kelimeler diğerlerine cevap veriyor?
- V: Value (değer) – bu kelimenin anlamı ne?
- dk: Key vektörlerinin boyutu (stabilite için bölünür)

### Çoklu Başlık Dikkat (Multi-Head Attention)

Transformer, dikkati farklı "başlıklarda" (head) paralel olarak hesaplar. Bu sayede farklı bağlam ilişkileri (örneğin gramatik, anlam, ilişki) aynı anda öğrenilir.

### Pozisyon Kodlaması (Positional Encoding)

Transformer kelime sırasını doğal olarak bilmez. Bu yüzden, her kelimeye **pozisyon bilgisi** eklenir. Bu bilgi, sinüs ve kosinüs fonksiyonlarıyla hesaplanan sabit bir desenle modellenir:

PE(pos,2i)=sin(100002i/dmodelpos)PE(pos,2i+1)=cos(100002i/dmodelpos)

Burada:

- pos: kelimenin dizideki pozisyonu
- i: vektör içindeki pozisyon
- dmodel: model boyutu (örn. 512)

Bu sayede model, "bugün hava güzel" ile "hava bugün güzel" arasındaki farkı anlayabilir.
